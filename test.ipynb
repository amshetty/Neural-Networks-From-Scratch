```python
import numpy as np 
import h5py 
import matplotlib.pyplot as plt
import math
import random

a=h5py.File("mnist_traindata.hdf5",'r') 
b=h5py.File("mnist_testdata.hdf5",'r') 
xdata_test=b["xdata"][:]/255
labels_test=b["ydata"][:]


xdata_train = a["xdata"][:]/255
labels_train= a["ydata"][:]
X=np.hstack((xdata_train,labels_train))
a=np.take(X,np.random.permutation(X.shape[0]),axis=0,out=X)

xdata_train = a[:,0:784]
labels_train = a[:,784:794]

X_train = xdata_train[0:50000,:]
X_val = xdata_train[50000:60000,:]
Y_train = labels_train[0:50000, : ]
Y_val = labels_train[50000:60000, :]
lr=0.005
acc=0
acc2=0

w1=np.random.rand(32,784)
b1=np.random.rand(32,500)

w2=np.random.rand(10,32) 
b2=np.random.rand(10,500)

def tanh2(x): 
    lst=[]  
    ans=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)) 
    return ans*ans


def tanh(x): 
    lst=[] 
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            ans=(np.exp(x[i][j]-np.mean(x[i]))-np.exp(-(x[i][j]-np.mean(x[i]))))/(np.exp(x[i][j]-np.mean(x[i]))+np.exp(-(x[i][j]-np.mean(x[i]))))
            lst.append(ans) 
    lst=np.asarray(lst).reshape(500,32) 
    return lst

def tanhgrad1(x):
    lst=[]
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            ans=1 - tanh2(x[i][j]-np.mean(x[i]))
            lst.append(ans)
    lst=np.asarray(lst).reshape(500,32)
    return lst

def smax(x): 
    lst=[] 
    for i in range(x.shape[0]):
        den=0
        for j in range(x.shape[1]):
            den=den+np.exp(x[i][j] - np.mean(x[i]))
        for j in range(x.shape[1]): 
            num=np.exp(x[i][j] - np.mean(x[i]))
            ans=(num/den) 
            lst.append(ans)
    return np.asarray(lst).reshape(500,10) 


for j in range(50):
    if j==15 or j==30:
        lr=lr/2
        
    acc=0
    print("Epoch: ",j)
    for i in range(0,59500,500):
        a1=tanh((w1@xdata_train[i:i+500].T +b1 ).T)
        a2=smax((w2@a1.T +b2).T)
        a2t=a2.T
        del2 = a2t-(labels_train[i:i+500].T)
        del1 = np.multiply((tanhgrad1(((w1@xdata_train[i:i+500].T) +b1).T)).T,(w2.T@del2))
        w1 = w1 - lr*(del1@xdata_train[i:i+500])
        b1 = b1 - lr*(del1)
        w2 = w2 - (lr*(del2@a1))
        b2 = b2 - lr*(del2)
        
        for k in range(500):
            if np.argmax(a2[k])==np.argmax(labels_train[i+k]):
                acc=acc+1
    acc=acc/60000
    print("Training accuracy is ",acc)
    
for i in range(0,9500,500):
    a111=tanh((w1@xdata_test[i:i+500].T + b1).T)
    a222=smax(((w2@a111.T) + b2).T)
    for k in range(500):
        if np.argmax(a222[k])==np.argmax(labels_test[i+k]):
            acc2=acc2+1
acc2=acc2/10000
    
print("Test accuracy is ", acc2)
```
